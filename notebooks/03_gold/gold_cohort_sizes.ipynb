```python
%sql
USE CATALOG etl_project;
USE SCHEMA bigquery_db_cohort_db;

-- Create Gold cohort size table: number of new customers per cohort month

CREATE OR REPLACE TABLE gold_cohort_sizes AS
SELECT
    cohort_month,
    COUNT(*) AS new_customers   -- total customers with first purchase in cohort month
FROM silver_customer_facts
GROUP BY cohort_month
ORDER BY cohort_month;
```
