```python
%sql
USE CATALOG etl_project;
USE SCHEMA bigquery_db_cohort_db;

-- Create Gold repeat purchase table: share of customers with 2+, 3+, 4+ total orders
-- Rounded to 4 decimals for analytical precision

CREATE OR REPLACE TABLE gold_repeat_purchases AS
SELECT
    cohort_month,
    
    -- % of customers with at least 2 lifetime orders
    ROUND(1.0 * COUNT_IF(total_orders >= 2) / COUNT(*), 4) AS repeat_rate_2_orders,
    
    -- % with at least 3 lifetime orders
    ROUND(1.0 * COUNT_IF(total_orders >= 3) / COUNT(*), 4) AS repeat_rate_3_orders,
    
    -- % with at least 4 lifetime orders
    ROUND(1.0 * COUNT_IF(total_orders >= 4) / COUNT(*), 4) AS repeat_rate_4_orders

FROM silver_customer_facts
GROUP BY cohort_month
ORDER BY cohort_month;
```
