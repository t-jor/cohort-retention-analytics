```python
%sql
USE CATALOG etl_project;
USE SCHEMA bigquery_db_cohort_db;

-- Create Gold monthly order distribution table:
-- Splits orders into new vs. existing customers per order month.

CREATE OR REPLACE TABLE gold_order_distribution AS

SELECT
    order_month,
    COUNT(*) AS total_orders,   -- all orders in the month

    -- orders from customers placing their first-ever order
    SUM(CASE WHEN is_first_order THEN 1 ELSE 0 END) AS orders_from_new_customers,

    -- orders from returning customers (second or later order)
    SUM(CASE WHEN is_repeat_order THEN 1 ELSE 0 END) AS orders_from_existing_customers
    
FROM silver_order_facts
GROUP BY order_month
ORDER BY order_month;
```
